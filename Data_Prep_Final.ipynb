```python
from google.colab import drive
from google.colab import files
import numpy as np
import os
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import confusion_matrix
import seaborn as sns

drive.mount('/content/drive')
myDrive = '/content/drive/Shareddrives/Deep_Learning_2024/Final/Data/Mixed_ETF/'
```

```python

file_names = ['vaw.us.txt','vcr.us.txt','vgt.us.txt','vfh.us.txt','vde.us.txt','vht.us.txt','vnq.us.txt']
data_files = {}
for name in file_names:
  file_path = os.path.join(myDrive,name)
  data_files[name] = pd.read_csv(file_path, sep=",", header=0)
```

# Check same dates for all stocks

```python
reference_df = data_files['vaw.us.txt']

min_length = min(df.shape[0] for df in data_files.values())
rows_to_check = min(3201, min_length)

for name in file_names:
    if name != 'vaw.us.txt':
        for i in range(rows_to_check):
            if data_files[name].loc[i, 'Date'] != reference_df.loc[i, 'Date']:
                print(f'we have a mismatch at row {i} in file {name}')
```

#Combine dataframes into a single one

```python
extended = data_files['vaw.us.txt'][['Date','Open', 'High', 'Low', 'Close', 'Volume']]


extended.columns = [f"{col}_vaw" for col in extended.columns]

for name in file_names:
    if name != 'vaw.us.txt':  #vaw alrady our starting point
        selected_cols = data_files[name][['Open', 'High', 'Low', 'Close', 'Volume']]
        #rename column adding the identifier
        selected_cols.columns = [f"{col}_{name[:3]}" for col in selected_cols.columns]

        # Concatenate into our extended dataframe
        extended = pd.concat([extended, selected_cols], axis=1)
```
